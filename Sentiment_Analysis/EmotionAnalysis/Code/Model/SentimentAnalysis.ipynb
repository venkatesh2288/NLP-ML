```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('combined_emotion.csv')
```

```python
import re
import nltk
from nltk.corpus import  stopwords
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
stopwords = stopwords.words('english')
nltk.download('wordnet')
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

lemma = WordNetLemmatizer()
stopwords_set = set(stopwords.words('english'))
corpus = []
for sentence in data['sentence']:
    sentence = re.sub('[^a-zA-Z]', ' ', sentence)
    sentence = sentence.lower().split()

    filtered_sentence = [lemma.lemmatize(word) for word in sentence if word not in stopwords_set]

    corpus.append(' '.join(filtered_sentence))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(corpus)
```

```python
x = tfidf_matrix
y = data['emotion']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 42 )
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
model = KNeighborsClassifier(n_neighbors=5)
model.fit(x_train, y_train)
```

```python
new_corpus = []
new_review = "i will rip your heart outta your body"
new_review = re.sub('[^a-zA-Z]', ' ', new_review)
new_review = new_review.lower()
new_review = new_review.split()
new_review = [lemma.lemmatize(word) for word in new_review if not word in set(stopwords.words('english'))]
new_review = ' '.join(new_review)
new_corpus.append(new_review)
new_corpus
```

```python
new_x = vectorizer.transform(new_corpus)
new_y = model.predict(new_x)
```
